```python
import numpy as np
import matplotlib.pyplot as plt

#     "adult.data"
data_path = "adult/adult.data"
data = np.genfromtxt(data_path, delimiter=',', dtype=str)

#    
age = []
education_num = []
hours_per_week = []

#    
for row in data:
    age.append(int(row[0]))  # 
    education_num.append(int(row[4]))  #   
    hours_per_week.append(int(row[12]))  #     

#  
plt.figure(figsize=(12, 4))

#  1:  vs   
plt.subplot(1, 3, 1)
plt.plot(age, education_num, 'bo', markersize=2)
plt.xlabel('Age')
plt.ylabel('Education Num')

#  2:  vs     
plt.subplot(1, 3, 2)
plt.plot(age, hours_per_week, 'ro', markersize=2)
plt.xlabel('Age')
plt.ylabel('Hours per Week')

#  3:    vs     
plt.subplot(1, 3, 3)
plt.plot(education_num, hours_per_week, 'go', markersize=2)
plt.xlabel('Education Num')
plt.ylabel('Hours per Week')

plt.tight_layout()
plt.show()
```
