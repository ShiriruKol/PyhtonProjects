```python
import numpy as np
import matplotlib.pyplot as plt

# Загрузка данных из файла "adult.data"
data_path = "adult/adult.data"
data = np.genfromtxt(data_path, delimiter=',', dtype=str)

# Извлечение данных из столбцов
age = []
education_num = []
hours_per_week = []

# Обход всей коллекции данных
for row in data:
    age.append(int(row[0]))  # возраст
    education_num.append(int(row[4]))  # количество лет обучения
    hours_per_week.append(int(row[12]))  # количество отработанных часов в неделю

# Строим графики
plt.figure(figsize=(12, 4))

# График 1: Возраст vs Количество лет обучения
plt.subplot(1, 3, 1)
plt.plot(age, education_num, 'bo', markersize=2)
plt.xlabel('Age')
plt.ylabel('Education Num')

# График 2: Возраст vs Количество отработанных часов в неделю
plt.subplot(1, 3, 2)
plt.plot(age, hours_per_week, 'ro', markersize=2)
plt.xlabel('Age')
plt.ylabel('Hours per Week')

# График 3: Количество лет обучения vs Количество отработанных часов в неделю
plt.subplot(1, 3, 3)
plt.plot(education_num, hours_per_week, 'go', markersize=2)
plt.xlabel('Education Num')
plt.ylabel('Hours per Week')

plt.tight_layout()
plt.show()
```
