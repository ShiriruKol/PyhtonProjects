```python
# https://chat.openai.com/c/458ea088-731a-43d1-ba58-a01f5b0fce67
#    
# https://archive.ics.uci.edu/dataset/19/car+evaluation
# Datasets 
import pandas as pd
import matplotlib.pyplot as plt

#    
data = pd.read_csv("car.data", header=None)

#   
data.columns = ["buying", "maint", "doors", "persons", "lug_boot", "safety", "class"]

#       
vhigh_data = data.iloc[:432]
high_data = data.iloc[432:864]
med_data = data.iloc[864:1296]
low_data = data.iloc[1296:]

#        
def plot_class_distribution(data, title):
    class_counts = data['class'].value_counts()
    class_counts.plot(kind='bar', color='skyblue')
    plt.title(title)
    plt.xlabel('')
    plt.ylabel('')
    plt.xticks(rotation=45)
    plt.show()

#        
plot_class_distribution(vhigh_data, '      "vhigh"')
plot_class_distribution(high_data, '      "high"')
plot_class_distribution(med_data, '      "med"')
plot_class_distribution(low_data, '      "low"')
```
