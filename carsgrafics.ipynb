```python
# https://chat.openai.com/c/458ea088-731a-43d1-ba58-a01f5b0fce67
# Какие запросы я делал
# https://archive.ics.uci.edu/dataset/19/car+evaluation
# Datasets 
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из файла
data = pd.read_csv("car.data", header=None)

# Назначение имен столбцам
data.columns = ["buying", "maint", "doors", "persons", "lug_boot", "safety", "class"]

# Создание подмножеств данных для каждого уровня покупки
vhigh_data = data.iloc[:432]
high_data = data.iloc[432:864]
med_data = data.iloc[864:1296]
low_data = data.iloc[1296:]

# Функция для отображения распределения уровней оценки класса автомобилей
def plot_class_distribution(data, title):
    class_counts = data['class'].value_counts()
    class_counts.plot(kind='bar', color='skyblue')
    plt.title(title)
    plt.xlabel('Класс')
    plt.ylabel('Количество')
    plt.xticks(rotation=45)
    plt.show()

# Построение распределения уровней оценки для каждого уровня покупки
plot_class_distribution(vhigh_data, 'Распределение уровней оценки для уровня покупки "vhigh"')
plot_class_distribution(high_data, 'Распределение уровней оценки для уровня покупки "high"')
plot_class_distribution(med_data, 'Распределение уровней оценки для уровня покупки "med"')
plot_class_distribution(low_data, 'Распределение уровней оценки для уровня покупки "low"')
```
